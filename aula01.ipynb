```python
data = {'ComDroga': [com_droga],
        'ComControle': [com_controle]
       }
  
df = pd.DataFrame(data,columns=['ComDroga','ComControle'])
print (df)
```

```python
data = {'Tratamento': ['Com Droga','Com Controle'],
        'Quantidade': [com_droga, com_controle]
       }
  
df = pd.DataFrame(data,columns=['Tratamento','Quantidade'])
df.plot(x ='Tratamento', y='Quantidade', kind = 'bar')
plt.show()
```
